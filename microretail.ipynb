```python
#Realizamos el diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15,8))
filtro_columna.plot(kind="box", vert=False)
plt.title("Valores atipicos del DataFrame")
plt.show() #Dibujamos el diagrama
```

```python
#Metodo aplicando desviacion estandar 
y=data
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers del dataframe
data1= data[(y>Limite_Superior)|(y<Limite_Inferior)]
data1
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el dataframe
data1= data[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data1
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Rellenamos las columas con 0, ya que son filtros que si no respondieron es porque no tienen.
data=data.fillna(0)
```

```python
#Corroboramos valores nulos
valores_lim=data.isnull().sum().sum()
valores_lim
```
