```python
#Corroboramos valores nulos en cada columna
valores_nulos=consumer.isnull().sum()
valores_nulos
```

```python
#Eliminamos columnas innecesarias, borre esta columna porque no en todas las tiendas van a vender los mismos productos
consumer= consumer.drop(["300_did_not_find"], axis=1)
consumer
```

```python
#Eliminamos columnas innecesarias, borre esta columna porque no en todas las tiendas van a tener los mismos productos adicionales, ademas tenemos pocos datos
consumer= consumer.drop(["284_additional_products_store"], axis=1)
consumer
```

```python
#Eliminamos columnas innecesarias, borre esta columna porque no en todas las tiendas van a tener los mismos productos adicionales
consumer= consumer.drop(["304_additional_services_store"], axis=1)
consumer
```

```python
#Corroboramos valores nulos en cada fila
valores_nulos=consumer.isnull().sum()
valores_nulos
```

```python
#Realizamos el diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15,8))
consumer.plot(kind="box", vert=False)
plt.title("Valores atipicos del DataFrame")
plt.show() #Dibujamos el diagrama
```

```python
#Metodo aplicando desviacion estandar 
y=consumer
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers del dataframe
outliers= consumer[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el dataframe
data3= consumer[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean.isnull().sum()
valores_nulos
```
